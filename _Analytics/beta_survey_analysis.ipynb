# Beta Survey Analysis

There are four items in the questionnaire that users can answer directly through options, which are
- Do you think the game's tutorial is clear enough?
- Compared to the side-view (2D) character, do you find the top-view (3D) character more difficult to control? Why or why not?
- Do you feel the game makes meaningful use of the 2D/3D switching mechanic?
- How easy was it for you to learn how to play the game?  

In addition, there are four other items are developmental questions that users can answer selectively, which are
- Describe the objective of the game in your own words/interpretations.
- What are some of the enjoyable aspects of the game so far? (What would you like to see more of)
- Describe any unclear or frustrating parts of the game.
- What improvements would you like to see?

## IMPORT LIBS

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

## Mandatory Questions Analysis

### 1. Do you think the game's tutorial is clear enough?

```python
file_path = 'Ctrl + Alt + Fun - Alpha Prototype Feedback Form Template (Responses).xlsx'
df = pd.read_excel(file_path)

column_index = 2
data = df.iloc[:, column_index].dropna().tolist()

counts = Counter(data)

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel(df.columns[column_index])
ax.set_ylabel('Count')
ax.set_title(f'Frequency of {df.columns[column_index]}')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### 2.   Compared to the side-view (2D) character, do you find the top-view (3D) character more difficult to control? Why or why not?

```python
column_index = 3
data = df.iloc[:, column_index].dropna().tolist()

def classify_answer(text):
    text = str(text).lower()
    if "both are about the same" in text:
        return "Both Same"
    elif "i'm not sure" in text or "im not sure" in text:
        return "Not Sure"
    elif "yes" in text and "3d" in text and "difficult" in text:
        return "3D Harder"
    elif "no" in text and "2d" in text and "difficult" in text:
        return "2D Harder"
    elif "3d" in text and any(word in text for word in ["frustrating", "annoying", "spamming", "jumping", "gravity", "cannot switch direction", "keep clicking"]):
        return "3D Annoying"
    elif "both are easy" in text:
        return "Both Easy"
    else:
        return "Other"

categorized_data = [classify_answer(x) for x in data]

counts = Counter(categorized_data)

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel('Answer Category')
ax.set_ylabel('Count')
ax.set_title('Character Control Difficulty')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### 3. Do you feel the game makes meaningful use of the 2D/3D switching mechanic?

```python
column_index = 4
data = df.iloc[:, column_index].dropna().tolist()

def simplify_answer(text):
    text = str(text).lower()
    if "yes" in text:
        return "Yes"
    else:
        return "No"

simplified_data = [simplify_answer(x) for x in data]

from collections import Counter
counts = Counter(simplified_data)

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 5))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel('Answer (Yes/No)')
ax.set_ylabel('Count')
ax.set_title('Simplified Responses')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### 4. How easy was it for you to learn how to play the game?

```python
column_index = 5
data = df.iloc[:, column_index].dropna().tolist()

counts = Counter(data)

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel(df.columns[column_index])
ax.set_ylabel('Count')
ax.set_title(f'Frequency of {df.columns[column_index]}')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Optional Questions

### 1. Describe the objective of the game in your own words/interpretations.

```python
column_index = 6
data = df.iloc[:, column_index].dropna().tolist()

def classify(text):
    text = text.lower()
    if "puzzle" in text or "solve" in text:
        return "Puzzle Solving"
    elif "switch" in text or "shift" in text or "perspective" in text:
        return "Switch 2D/3D"
    elif "clear" in text or "complete level" in text:
        return "Clear Levels"
    elif "trap" in text or "avoid" in text:
        return "Avoid Obstacles"
    elif "portal" in text:
        return "Use Portals"
    elif "jump" in text:
        return "Jumping Game"
    elif "not sure" in text:
        return "Not Sure"
    else:
        return "Reach Flag"

categories = [classify(x) for x in data]

counts = Counter(categories)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel('Category')
ax.set_ylabel('Count')
ax.set_title('Summary of Game Objectives')
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### 2. What are some of the enjoyable aspects of the game so far? (What would you like to see more of)

```python
column_index = 7
data = df.iloc[:, column_index].dropna().tolist()

def classify(text):
    text = str(text).lower()
    if "switch" in text or "perspective" in text or "change view" in text:
        return "2D/3D Switching"
    elif "puzzle" in text or "solve" in text:
        return "Puzzle Solving"
    elif "box" in text or "carry" in text or "picking" in text or "placing" in text:
        return "Box Mechanics"
    elif "strategy" in text or "strategize" in text:
        return "Strategy Thinking"
    elif "maze" in text or "level design" in text:
        return "Level Design"
    elif "jump" in text:
        return "Jumping"
    elif "none" in text or "n/a" in text:
        return "None"
    else:
        return "Other"

categories = [classify(x) for x in data]

counts = Counter(categories)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel('Category')
ax.set_ylabel('Count')
ax.set_title('Summary of Game Objectives')
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

### 3. Describe any unclear or frustrating parts of the game.

```python
column_index = 8
data = df.iloc[:, column_index].dropna().tolist()

def classify(text):
    text = str(text).lower()
    if "portal" in text:
        return "Portal Usage Unclear"
    elif "tutorial" in text:
        return "Tutorial Unclear"
    elif "switch" in text or "view" in text or "perspective" in text:
        return "2D/3D Switching Unclear"
    elif "hole" in text or "trap" in text or "fall" in text:
        return "Hole/Trap Confusion"
    elif "movement" in text or "control" in text or "arrow key" in text:
        return "Movement Control Issue"
    elif "placement" in text or "drop block" in text:
        return "Block Placement Difficulty"
    elif "not clear" in text or "unclear" in text or "confusing" in text:
        return "General Unclarity"
    elif "na" in text or "none" in text or 'n/a' in text:
        return "None"
    else:
        return "Other"

categories = [classify(x) for x in data]

counts = Counter(categories)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(counts.keys(), counts.values())

ax.set_xlabel('Category')
ax.set_ylabel('Count')
ax.set_title('Summary of Game Objectives')
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
